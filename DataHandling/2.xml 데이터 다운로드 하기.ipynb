기상청 데이터 다운로드 하기
RSS 서비스 - 텍스트로 축약해 놓은것
http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp

http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=133
충청남도의 데이터만 받고싶을때 (133)

- get 방식(Web에서) : url을 통해 데이터 전송

stnId에 133이라는 값을 전달한것 (지역마다 할당한 번호,전국 108, 서울경기 109)

?를 통해 변수임을 알림

다른변수 추가할때는 &로 구분

?stnId = 133 & a = aaaa

서버에서 저걸 읽어서 정해진 대로 동작하는 것임

```python
import urllib.request as ur
import urllib.parse #url을 인코딩 하기위해 불러오는 모듈(파싱작업을 해줘야함)

rssUrl = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

#매개변수 : 지역별 코드를 지정하는 변수
"""
stnId 코드
-> 108:전국, 109:서울/경기도. 105:강원도, 131:충북
    133: 충남, 146:전북, 156:전남, 143:경북
    159:경남,184:제주도
"""
values = {
    'stnId':'146',
}

params = urllib.parse.urlencode(values) #인코딩 해줌??

url = rssUrl + '?'+params
data = urllib.request.urlopen(url).read()
text = data.decode('utf-8') #url 데이터를 읽을수 있도록 디코드 해야함

print(text)
```

```python
"""
위같이 하면 지역 바뀔때마다 값을 바꿔줘야하는데
실행시 아예 매개변수를 받으면?
ex)
>>python ex03.py 146

sys 모듈을 통해 명령줄 인수 받아올수 있음
(sys.argv) -> argr[0~] list 로 받아올수 있음
"""
import sys
import urllib.request as ureq
import urllib.parse as upar 

#명령줄 인수가 제대로 입력되었는 확인
if len(sys.argv) <= 1: #명령줄 인수가 1이하이면 오류 메시지 출력
    print("사용법 : >>python 인수1 인수2")
    sys.exit()
    
localNum = sys.argv[1]

rssUrl = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

values = {
    'stnId':str(localNum)
}

params = upar.urlencode(values)
url = rssUrl + '?'+params

print(url)

#RSS 데이터를 다운로드
data = ureq.urlopen(url).read()
text = data.decode('utf-8')
print(text)
```
